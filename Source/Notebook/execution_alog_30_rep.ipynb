```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
### Binary Trees

# Noms des fichiers
file_names = [ "../Algo/BinaryTrees/C/binary_trees/serveur19_c_execution_results.csv",
              "../Algo/BinaryTrees/C/binary_trees/serveur19_c-o1_execution_results.csv",
              "../Algo/BinaryTrees/C/binary_trees/serveur19_c-o2_execution_results.csv",
              "../Algo/BinaryTrees/C/binary_trees/serveur19_c-o3_execution_results.csv",
              "../Algo/BinaryTrees/C++/binary_trees/serveur19_cpp_execution_results.csv",
              "../Algo/BinaryTrees/C++/binary_trees/serveur19_cpp-o1_execution_results.csv",
              "../Algo/BinaryTrees/C++/binary_trees/serveur19_cpp-o2_execution_results.csv",
              "../Algo/BinaryTrees/C++/binary_trees/serveur19_cpp-o3_execution_results.csv",
              "../Algo/BinaryTrees/Erlang/binary_trees/serveur19_erlang_execution_results.csv",
              "../Algo/BinaryTrees/Fortran/binary_trees/serveur19_fortran_execution_results.csv",
              "../Algo/BinaryTrees/Fortran/binary_trees/serveur19_fortran-o1_execution_results.csv",
              "../Algo/BinaryTrees/Fortran/binary_trees/serveur19_fortran-o2_execution_results.csv",
              "../Algo/BinaryTrees/Fortran/binary_trees/serveur19_fortran-o3_execution_results.csv",
              "../Algo/BinaryTrees/Java/binary_trees/src/serveur19_java_execution_results.csv",
              "../Algo/BinaryTrees/Python/binary_trees/serveur19_python_execution_results.csv",
              "../Algo/BinaryTrees/Ruby/binary_trees/serveur19_ruby_execution_results.csv",
              "../Algo/BinaryTrees/Rust/binary_trees/serveur19_rust_execution_results.csv",
              "../Algo/BinaryTrees/Rust/binary_trees/serveur19_rust-o1_execution_results.csv",
              "../Algo/BinaryTrees/Rust/binary_trees/serveur19_rust-o2_execution_results.csv",
              "../Algo/BinaryTrees/Rust/binary_trees/serveur19_rust-o3_execution_results.csv",
              ]

# Initialiser un dictionnaire pour stocker les moyennes des temps d'excution rels
moy_time_reals = {}
moy_systs = {}
moy_users= {}

for fichier in file_names:
    # Extraire le nom du langage  partir du nom de fichier
    langage = fichier.split('_')[2]
    
    # Lire le fichier CSV
    df = pd.read_csv(fichier, sep=',', names=['User Time (s)', 'System Time (s)', 'Real Time (s)', 'Executable Size (KB)'])
    
    # Real Time
    df['Real Time (s)'] = pd.to_numeric(df['Real Time (s)'].iloc[1:], errors='coerce')
    moy_time_real = df['Real Time (s)'].mean()
    moy_time_reals[langage] = moy_time_real

    # Syst Time
    df['System Time (s)'] = pd.to_numeric(df['System Time (s)'].iloc[1:], errors='coerce')
    moy_syst = df['System Time (s)'].mean()
    moy_systs[langage] = moy_syst

    # User Time
    df['User Time (s)'] = pd.to_numeric(df['User Time (s)'].iloc[1:], errors='coerce')
    moy_user = df['User Time (s)'].mean()
    moy_users[langage] = moy_user

# Real Time graph
plt.figure(figsize=(12, 8))
languages = list(moy_time_reals.keys())
tps_reel = list(moy_time_reals.values())
plt.bar(languages, tps_reel, color='skyblue')
plt.title('Average Real Execution Times For Binary Tree Traversal Across Programming Languages')
plt.xlabel('Programming Language')
plt.ylabel('Average Real Time (s)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.savefig("real_time_binary_tree.pdf")
plt.show()

# User Time graph
plt.figure(figsize=(12, 8))
languages = list(moy_users.keys())
tps_user = list(moy_users.values())
plt.bar(languages, tps_user, color='skyblue')
plt.title('Average USer Execution Times For Binary Tree Traversal Across Programming Languages')
plt.xlabel('Programming Language')
plt.ylabel('Average Real Time (s)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.savefig("user_time_binary_tree.pdf")
plt.show()

# Syst Time graph
plt.figure(figsize=(12, 8))
languages = list(moy_systs.keys())
tps_syst = list(moy_systs.values())
plt.bar(languages, tps_syst, color='skyblue')
plt.title('Average Syst Execution Times For Binary Tree Traversal Across Programming Languages')
plt.xlabel('Programming Language')
plt.ylabel('Average Real Time (s)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.savefig("syst_time_binary_tree.pdf")
plt.show()

```
